```python
import pickle 
import numpy as np 
import pandas as pd 

car_2019 = pd.read_csv('car_2019.csv')
car9=car_2019.drop(['End_Lat', 'End_Lng','Source'], axis=1)
car9.dropna(subset = ["Number","Precipitation(in)","Wind_Speed(mph)","Wind_Direction","Visibility(mi)","Pressure(in)","Humidity(%)","Temperature(F)","Wind_Chill(F)","Severity"], inplace=True)
car9 = car9.reset_index(drop=True)
car9_weather=car9.drop(['ID','TMC','Start_Time','End_Time','Start_Lat','Start_Lng','Distance(mi)','Description','Number','Street','Side','City','County','State','Zipcode','Country','Timezone','Airport_Code','Weather_Timestamp','Wind_Direction','Weather_Condition','Amenity','Bump','Crossing','Give_Way','Junction','No_Exit','Railway','Roundabout','Station','Stop','Traffic_Calming','Traffic_Signal','Turning_Loop','Sunrise_Sunset','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight'],axis=1)

loaded_model = pickle.load(open('knn2017model', 'rb'))

import matplotlib.pyplot as plt
from pandas import DataFrame
import numpy as np
```

# validating model with 2018 &2019 data for Top 3 Features (windchill/precipitation/humidity)

```python
f1='Wind_Chill(F)'
f2='Precipitation(in)'
f3='Humidity(%)'

accuracy=[]

for i in range(50):

    car = car9_weather.sample(frac=0.25)

    y_test=car.iloc[:,0]
    y_test

    X_test=car[[f1,f2,f3]]
    X_test

    #loaded_model = pickle.load(open('knnpickle_file', 'rb'))
    result = loaded_model.score(X_test, y_test)
    accuracy.append(result)

```

```python
df = DataFrame(accuracy,columns=['accuracy'])
x = df.index
y = df['accuracy']   

# Calculate the simple average of the data
y_mean = [np.mean(y)]*len(x)
fig,ax = plt.subplots()

# Plot the data
data_line = ax.plot(x,y, label='Data', marker='o')

# Plot the average line
mean_line = ax.plot(x,y_mean, label='Mean', linestyle='--')

# Make a legend
legend = ax.legend(loc='upper right')

plt.plot(df.index,df['accuracy'])
plt.title('Validation with 2019 Data for Top 3 Features (windchill/precipitation/humidity)')
plt.xlabel('trial')
plt.ylabel('accuracy')

plt.show()
```
